```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
filename = 'MAR_SP.xlsx'
filename_new = 'MAR_K1.xlsx'

book = load_workbook(filename)
sheet = book['Сложные топонимы']
sheet1 = book['Компоненты 1']

print(sheet)
```

```python
for index, row in enumerate(sheet, start=1):
    reg = sheet[f'a{index}'].value
    name = sheet[f'b{index}'].value
    otype = sheet[f'c{index}'].value
    name_parts = name.split()
    first_part = name_parts[0]
    after_part0 = name.replace(f'{first_part} ', '')
#    print(after_part0)
    after_part = after_part0.strip(')(')
    
#    print('1', morph.parse(first_part)[0].tag.POS, first_part)
#    print('2', morph.parse(after_part)[0].tag.POS, after_part)

    sheet1[f'a{index}'].value = reg
    sheet1[f'b{index}'].value = first_part
    sheet1[f'c{index}'].value = after_part
    main_part = ''
    if len(after_part.split()) > 1:
        rem_parts = after_part.split()
        if morph.parse(rem_parts[-1])[0].tag.POS != 'NOUN' and morph.parse(rem_parts[0])[0].tag.POS == 'NOUN':
            main_part = rem_parts[0]
        elif morph.parse(rem_parts[-1])[0].tag.POS == 'NOUN' and morph.parse(rem_parts[0])[0].tag.POS != 'NOUN':
            main_part = rem_parts[-1]
        else:
            main_part = rem_parts[-1]
    
    else:
        main_part = after_part
    
    sheet1[f'd{index}'].value = main_part
    sheet1[f'e{index}'].value = otype
```

```python
book.save(filename_new)
```
